```python
import pandas as pd
df=pd.read_csv("tennis.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
outlook=LabelEncoder()
humidity=LabelEncoder()
windy=LabelEncoder()
play=LabelEncoder()
```

```python
df['outlook']=outlook.fit_transform(df['outlook'])
df['humidity']=outlook.fit_transform(df['humidity'])
df['windy']=outlook.fit_transform(df['windy'])
df['play']=outlook.fit_transform(df['play'])
```

```python
df
```

```python
x=df[['outlook','humidity','windy']]
y=df.play
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='gini') #gini is by default too
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.predict(x_test)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(classifier)
```
