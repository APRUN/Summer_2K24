# Logistic Regression (Binary Classification)




```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("TermInsuranceData3.csv")
df
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,1]
y
```

```python
plt.scatter(x,y)
```



This 'S' shaped line is called **Sigmoid function.** It actually divides the y-axis into probability ranging from 0->1 and make a cut line. The points are decided in basis of the cut line.

```python
df['Bought Insurance'].replace({'no':'0','yes':'1'},inplace=True) #inplace used to save in the dataset permanently
df
```

```python
plt.scatter(x='Age',y='Bought Insurance',data=df)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(df[['Age']],df['Bought Insurance'],test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

 # Logistic Regression (MultiClass Classification)
 
 

```python
import pandas as pd
df=pd.read_csv("IrisDataset.csv")
df
```

```python
df['Species'].unique()
```

```python
df['Species'].replace({'iris-virginica':'1', 'iris-setosa':'2', 'iris-versicolor':'3'},inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']],df['Species'],test_size=0.2)
```

```python
len(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
x_test
```

```python
lr.score(x_test,y_test)
```
