# Machine Learning:
*The science of getting computers to learn and act like humans do, and improve their learning over time in autonomus fashion by feeding them data and information in the form of observations and real world interactions.*

## Working:


## Characteristics of ML:
1. The potential to perform automated data visualization.
2. Precise Data Analysis
3. Business intelligence at its finest
## Advantages:
1. Resolving Complex problems
2. Automation of tasks
3. Trends and patterns activation
4. Wide range of applications

## Disadvantages:
1. Data Acquisition (Data could be incorrect used for training)
2. Highly Error prone
3. Algorithm Selection
4. Time Consuming

## Classification:
1. **Supervised Learning:** The data provided is labeled. It is under a supervisor.
2. **Unsupervised Learning:** No supervisor, Unlabeled Data. Machine explores itself creates pattern to understand.
3. **Reinforcment Learning:** Machine learns itself. An agent is used that gets feedback each times in form of penalty and reward.

## Life Cycle:
1. **Data Collection:**
    1. Identify various source Information
    2. Gather Data
    3. Combine data acquired from data source 
3. **Data Prepairation:** Can be referred as data eploration. It means to explore or study your data
4. **Data Wrangling:** Its a process of cleaning and converting raw data into useful information.
   1. Cleaning up raw data
   2. Filtering Noise
   3. Recognizing and removing outliners(abnormal readings)
   4. Removing or filling missing data
6. **Data Modeling:** It is the step in which we take the data and select a ML algorithm to build a model.
    1. Select ML algorithm
    2. Build the model
    3. validate the result
8. **Model Training:** The Model is trained by giving data and data should be provided sufficient to get the model trained.
9. **Model Testing:** Also known as **Model verification** which is used to check for the accuracy of model by providing with inputs that are unseen.
10. **Deployment:** Ready for production.

## Dataset:
"A collection of records usually presented in tabular form."

## Types of Dataset:
1. **Numeric(Quantitative) Data:** Data is in numbers.
2. **Categorical(Qualitative) Data:** Data is categorized into types like true/false, male/female, 0/1 etc
3. **Ordinal Data:** Hybrid of above two types. Like giving 5 star reviews over the app but it actually means 'Excellent'.
