# Types of Machine Learning:
## Supervised Learning:
In supervised machine learning, you train the machine using data which is well labeled , i.e. some input data is already tagged with correct anwser and his algorithm learns from labeled training data that helps you to predict the further outcomes.

## Workng:


## Types of Supervised ML:
1. **Regression:** Regression is numeric. It is used where there is a relationship between the dependent and independent variable and where the values are continous. Like determining salaries and determining weight etc.
    1. Linear Regression
    2. Polynomial regression
    3. Regression Tree
3. **Classification:** Classification is categorical. e.g for true/false, male/female. If the labels are divided into two different classes we call it **Binary Classification.** Else more than two called **Multi-class Classification.**
    1. Random forest
    2. Deciscion Tree
    3. Logistic regression
**Advantages of Supervised Learning:**
1. You have full control over the machine learning model coz we are providing the dataset.
2. You can easily debug and find errors.
3. You can determine the number of classes.
   
**DisAdvantages of Supervised Learning:**
1. Have Limited Scope
2. Collecting labelled data is expensive
3. Wrong Predictions for unknown classes
